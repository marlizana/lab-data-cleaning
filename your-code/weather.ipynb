## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
weather = pd.read_csv("../weather-raw.csv")
```

```python
#Hacemos una primera visualización de la tabla. Vemos que disponemos de muy pocos datos.
weather
```

```python
#Sacamos unos primeros datos agregados en los que vemos que varios de los sensores de temperatura.
"""
Vemos como varios de los días dentro del mes no disponen de datos del año 
"""
stats = weather.describe().transpose()
stats
```

```python
null_cols = weather.isnull().sum()
```

```python
#Con esto eliminamos las columnas que solo tienen NaN.
drop_cols = list(null_cols[null_cols >= 22].index)
df = weather.drop(drop_cols, axis = 1)
df
```
